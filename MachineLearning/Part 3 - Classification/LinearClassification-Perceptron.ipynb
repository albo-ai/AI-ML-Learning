```python
import matplotlib.pyplot as plt
%matplotlib notebook
import random
```

```python
plus_list = []
train_labels = []
minus_list = []
labels_plus_list = []
labels_minus_list = []
test_plus_list = []
test_labels_plus_list = []
test_minus_list = []
test_labels_minus_list = [] 

for x in range(0, 20):
    for y in range(0, 20):
        if((y > x) and random.random() > 0.7):
            plus_list.append(tuple((x,y)))
            labels_plus_list.append('+')
        if((y < x) and random.random() > 0.7):
            minus_list.append(tuple((x,y)))
            labels_minus_list.append('-')
        if((y > x) and random.random() < 0.7):
            test_plus_list.append(tuple((x,y)))
            test_labels_plus_list.append('+')
        if((y < x) and random.random() < 0.7):
            test_minus_list.append(tuple((x,y)))
            test_labels_minus_list.append('-')
            

#print(plus_list)
#print(minus_list)
train_data = plus_list + minus_list
train_labels = labels_plus_list + labels_minus_list
test_data = test_plus_list + test_minus_list
test_labels = test_labels_plus_list + test_labels_minus_list
#print(train_data)
print(train_labels)
```

```python
plt.scatter(*zip(*plus_list),s=None,c=None,marker="+")
plt.scatter(*zip(*minus_list),s=None,c=None,marker="_")
```

```python
import sklearn.linear_model as sk
clf = sk.Perceptron()
clf.fit(train_data,train_labels)

print('accuracy:', clf.score(train_data, train_labels))
print('accuracy:', clf.score(test_data, test_labels))
```

```python
import numpy as np

colormap = np.array(['r', 'k'])
# Plot the original data
#plt.scatter(train_data, train_labels, c=colormap[train_labels], s=40)
#plt.scatter(*zip(*train_data),s=None,c=colormap[train_labels])
 
# Output the values
#print "Coefficient 0 " + str(clf.coef_[0,0])
#print "Coefficient 1 " + str(clf.coef_[0,1])
#print "Bias " + str(clf.intercept_)
plt.scatter(*zip(*plus_list),s=None,c=None,marker="+")
plt.scatter(*zip(*minus_list),s=None,c=None,marker="_")
plt.scatter(*zip(*test_plus_list),s=None,c=None,marker="*")
plt.scatter(*zip(*test_minus_list),s=None,c=None,marker="o")

# Calc the hyperplane (decision boundary)
ymin, ymax = plt.ylim()
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(ymin, ymax)
yy = a * xx - (clf.intercept_[0]) / w[1]
 
# Plot the line
plt.plot(yy,xx, 'k-')
plt.show()
```
