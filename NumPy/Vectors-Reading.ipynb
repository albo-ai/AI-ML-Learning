A Vector has a magnitude (size) and Direction.<br>
A Vector **x** is denoted as  : $\vec{\mathbf{x}}$

* Length of the line shows its **magnitude**.
* Arrow head points in the **direction**.

Ex: Velocity, Acceleration, Force, etc. are Vectors.<br>

Vector in 1-D --> simply a real number<br>
Vector in 2-D --> represents a point on a plane<br>
$\vec{\mathbf{a}}$= (2, 3)<br>
$\vec{\mathbf{b}}$ = (4, 5)<br>

Vectors have a dimension.<br>
An d-Dimensional vector is represented as :<br>
$$
\vec{\mathbf{u}}=
[u_1, u_2, ....., u_d]
$$
Individual elements are indexed by a subscript.
***
__Addition of Vectors__<br>
Two vector  $\vec{\mathbf{u}}$ and  $\vec{\mathbf{v}}$ haivng same dimension 'n' can be added to get a vector  $\vec{\mathbf{w}}$ =  $\vec{\mathbf{u}}$ +  $\vec{\mathbf{v}}$ which also has the same dimension 'n'. It's something like, *co-ordinate wise addition*.

$$
w_i = u_i + v_i \quad \textrm{ for all i = 1 to n}
$$
$\vec{\mathbf{a}}$= (2, 3)<br>
$\vec{\mathbf{b}}$ = (4, 5)<br>
$\vec{\mathbf{c}}$= $\vec{\mathbf{a}}$ + $\vec{\mathbf{b}}$= (2 + 4, 3 + 5) = (6, 8)<br><br>
What we are doing is : Adding __x__ parts & adding __y__ parts.<br>
$$
\vec{\mathbf{a}}= (x_1, y_1)\\
\vec{\mathbf{b}} = (x_2, y_2)\\
\vec{\mathbf{c}} = (x_1 + x_2, y_1 + y_2)
$$

***
__Scalar Multiplication__<br>
It is like, multiplying a vector with scalar value. <br>
You can understanding this, as "scaling" a vector. It could be either making it big or small. <br>
$$
\vec{\mathbf{a}} = (2, 3, 5) \\
3 * \vec{\mathbf{a}} = (6, 9, 15)
$$
However, please note that the Direction of the vector remains same.<br>
***
__Norm or Length of Vector or Magnitude of Vector__<br>
Some times, we may also say it as L2 Norm.<br>
* The objective is to measure the length of a Vector. Please remember, this is going to be always a +ve number.<br>
$$
\vec{\mathbf{a}} = (6, 8) \\
\mid  \mid a \mid \mid \ = \sqrt{6^2 + 8^2} = \sqrt{36 +  64} = 10
$$
<br>L2 Norm (or) Length of Vector (or) Magnitude of Vector :
$$
\mid  \mid v \mid \mid \ = \sqrt{v_1^2 + v_2^2 + .... + v_d^2}
$$
<br>
* A Vector with Norm 1 (or) length of a vector is 1 is called __ Unit Vector__

#### Triangle Inequality :
$$
\mid  \mid u + v \mid \mid \ \leq \ \mid  \mid u \mid \mid \ + \mid  \mid v \mid \mid
$$
<br>
This will be equal when u and v are __collinear vectors__. <br>
$$
\vec{\mathbf{u}} \ = c . \vec{\mathbf{v}} \ , c > 0
$$
In the above case, u and v are called collinear vectors.<br>
***
#### Unit Vector 
It is also some times called as Direction Vector.<br>
* A Unit Vector is of length or magnitude 1.

$$
\hat v =  \frac{v}{\mid  \mid v \mid \mid}
$$
