```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize']=10,8
```

```python
excelDF = pd.read_excel("C:\\SuperStore.xls", sheet_name = "Orders")
```

```python
region = excelDF.Region.unique()
for r in region:
    subset = excelDF[excelDF.Region == r]
    sns.distplot(subset['Quantity'],hist=False,kde=True,label=r,kde_kws = {'linewidth': 3})
plt.ylabel('Density')
```

```python

sns.distplot(excelDF['Quantity'],hist=True,kde=True,kde_kws = {'shade':True,'linewidth': 3,"color":"Red"},hist_kws={'edgecolor':'black'},bins=10)
plt.ylabel('Density')
```

```python
# j = sns.jointplot(data=movies, x='CriticsRatings',y='AudienceRatings',color='Red',kind="hex")
j = sns.jointplot(data=excelDF, x='Quantity',y='Profit',color='Red',kind="hex")
j.annotate(stat=None)
```
