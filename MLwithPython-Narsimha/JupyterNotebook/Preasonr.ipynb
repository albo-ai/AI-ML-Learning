```python
import seaborn as sns
import pandas as pd
dicttest= { 'x':[1,2,3,4,5],'y':[2,4,6,8,10]}
# dicttest
df = pd.DataFrame(dicttest,columns=['x','y'])
vis3 = sns.lmplot(data = df, x="x",y="y",fit_reg=False)
```
