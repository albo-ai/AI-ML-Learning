```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
plt.rcParams["figure.figsize"]=8,10
plt.rcParams["patch.force_edgecolor"] = True
movies = pd.read_csv("C:\\Users\\Bibhu\\Desktop\\Oracle_ML\\Assignment\\Movie-Ratings.csv")
```

```python

movies.columns = ['Film','Genre','CriticRatings','AudienceRatings','BudgetMillions','Year']
movies["Genre"]=movies.Genre.astype('category')
movies["Year"]=movies.Year.astype('category')
```

```python
# histogram using distplot

sns.distplot(movies.CriticRatings)
plt.show()
```

```python
# histogram using pyplot
sns.set_style("darkgrid")  # this changes the background to grid
#sns.set_style("white")
plt.hist(movies.CriticRatings,bins=10)
plt.show()
```

```python
plt.hist(movies.AudienceRatings,bins=10)
plt.show()
```

# Stacked Histogram

```python
plt.hist(movies[movies.Genre == "Action"].BudgetMillions, stacked=True)
plt.hist(movies[movies.Genre == "Romance"].BudgetMillions,stacked=True)
plt.hist(movies[movies.Genre == "Drama"].BudgetMillions,stacked=True)
plt.hist(movies[movies.Genre == "Thriller"].BudgetMillions,stacked=True)
plt.show()
```

```python
# manual stacking of histogram
plt.hist([movies[movies.Genre == "Action"].BudgetMillions, 
movies[movies.Genre == "Romance"].BudgetMillions,
movies[movies.Genre == "Drama"].BudgetMillions,
movies[movies.Genre == "Thriller"].BudgetMillions],bins=15,stacked=True)
plt.show()
```

```python
plt.rcParams["figure.figsize"]=8,4
budgetList = []  # empty list creation
genreLabel = []
colorList = ["Red","Blue","Green","Yellow","Pink","Black","Magenta"]
for gen in movies.Genre.cat.categories:
    genreLabel.append(gen)
    budgetList.append(movies[movies.Genre == gen].BudgetMillions)

plt.hist(budgetList,bins=30,stacked=True,label=genreLabel,color=colorList)
plt.legend()
plt.show()
```
