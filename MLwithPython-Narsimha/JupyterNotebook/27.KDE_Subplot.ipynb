```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
plt.rcParams["figure.figsize"]=8,10
plt.rcParams["patch.force_edgecolor"] = True
```

```python
movies = pd.read_csv("C:\\Users\\Bibhu\\Desktop\\Oracle_ML\\Assignment\\Movie-Ratings.csv")
movies.columns = ['Film','Genre','CriticRatings','AudienceRatings','BudgetMillions','Year']
movies["Genre"]=movies.Genre.astype('category')
movies["Year"]=movies.Year.astype('category')
```

```python
# drawing a regular scatter plot
sns.set_style("darkgrid")
sns.lmplot(data=movies,x='CriticRatings',y='AudienceRatings', hue ="Genre", fit_reg=False, size = 7)
plt.show()
```

```python
# plot the KDE plot
sns.kdeplot(movies.CriticRatings,movies.AudienceRatings)
plt.show()
```

```python
# plot the KDE plot styling
sns.kdeplot(movies.CriticRatings,
            movies.AudienceRatings,
            shade=True)
plt.show()
```

```python
# shade_lowest=False gets the grid back
sns.kdeplot(movies.CriticRatings,movies.AudienceRatings,shade=True,shade_lowest=False)
plt.show()
```

```python
# change the plot color
sns.kdeplot(movies.CriticRatings,movies.AudienceRatings
            ,shade=True,shade_lowest=False,
            cmap='Reds')
plt.show()
```

```python
# get a smoother outline for the plotting call the kdeplot twice
sns.kdeplot(movies.CriticRatings,movies.AudienceRatings,
            shade=True,shade_lowest=False,cmap='Reds')
sns.kdeplot(movies.CriticRatings,movies.AudienceRatings,
            cmap='Reds')
plt.show()
```
