# Filtering Rows in DataFrame

```python
# filtering rows
df.InternetUsers < 2
```

```python
# filter1 with boolean values against rows satisfying the condition
filter1 = df.InternetUsers < 2
```

```python
# giving the filter1 object to dataframe , it returns all rows where True was mentioned
df[filter1]
```

```python
# rather than creating a filter we can use the condition directly in the dataframe

df[df.InternetUsers < 2]

df[df["Internet users"] < 2]
```

```python
filter2 = df.BirthRate > 40
```

```python
df[filter2]
```

```python
# error
df[filter1 and filter2]
```

```python
# logica and
df[filter1 & filter2]
```

```python
# logical or
df[(filter1 | filter2)]
```

```python
# not equal to
#df[filter1 != filter2]

# mention brackets , compulsory

df[(df.InternetUsers < 2) & (df.BirthRate > 40)]
```

# update values in dataframe .at and .iat

```python
# .at refers to the label for both row and column   
# .at[10, 'BirthRate']     10 is the index number not index position 10, and BirthRate is the column label
```

```python
# creating a dataframe
df2 = df[::10]
```

```python
# look at the indexing
df2
```

```python
# 10 is index label 10 (which in real is 2nd row)
df2.at[10,'CountryName']
```

```python
# .iat     (refers to values at actual index potions and not the labels for both rows and columns, no column names allowed)

df2.iat[10,0]

#df2.head()
```
